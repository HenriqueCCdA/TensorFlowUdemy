```python
from sklearn.preprocessing import MinMaxScaler
scaler_x = MinMaxScaler()

base[
    [
        'bedrooms',
        'bathrooms',
        'sqft_living',
        'sqft_lot',
        'floors',
        'waterfront',
        'view',
        'condition',
        'grade',
        'sqft_above',
        'sqft_basement',
        'yr_built',
        'yr_renovated',
        'zipcode',
        'lat',
        'long'
    ]
] = scaler_x.fit_transform(
    base[
        [
            'bedrooms',
            'bathrooms',
            'sqft_living',
            'sqft_lot',
            'floors',
            'waterfront',
            'view',
            'condition',
            'grade',
            'sqft_above',
            'sqft_basement',
            'yr_built',
            'yr_renovated',
            'zipcode',
            'lat',
            'long'
        ]
    ]
)
```

```python
base.head()
```

```python
scaler_y = MinMaxScaler()
base[['price']] = scaler_y.fit_transform(base[['price']])
```

```python
base.head()
```

```python
X = base.drop('price', axis=1)
y = base.price
```

```python
previsoes_colunas = colunas_usadas[1:17]
previsoes_colunas
```

```python
import tensorflow as tf
tf.random.set_seed(42)
tf.__version__
```

```python
colunas = [tf.feature_column.numeric_column(key=c) for c in previsoes_colunas]
```

```python
colunas
```

```python
from sklearn.model_selection import train_test_split
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_treinamento.shape
```

```python
y_treinamento.shape
```

```python
X_teste.shape
```

```python
def make_input_fn(X, y = None, num_epochs=None, shuffle=True, batch_size=32):
    def input_function():
        ds = tf.data.Dataset.from_tensor_slices((dict(X), y))
        if shuffle:
            ds = ds.shuffle(1_000)
        ds = ds.batch(batch_size).repeat(num_epochs)
        return ds
    return input_function
```

```python
ds = tf.data.Dataset.from_tensor_slices((dict(X_treinamento), y_treinamento))
```

```python
funcao_treinamento = make_input_fn(X_treinamento, y_treinamento)
```

```python
funcao_treinamento()
```

```python
funcao_teste = make_input_fn(X_teste, y_teste, num_epochs=10_000, shuffle=False)
```

```python
config = tf.estimator.RunConfig(tf_random_seed=42)
regressor = tf.estimator.LinearRegressor(feature_columns=colunas, config=config)
```

```python
regressor.train(input_fn=funcao_treinamento, steps=10_000)
```

```python
metricas_treinamento = regressor.evaluate(input_fn=funcao_treinamento, steps=10_000)
```

```python
metricas_teste = regressor.evaluate(input_fn=funcao_teste, steps=10_000)
```

```python
metricas_treinamento
```

```python
metricas_teste
```

```python
funcao_previsao = make_input_fn(X_teste, num_epochs=1, shuffle=False)
```

```python
previsoes = regressor.predict(input_fn=funcao_previsao)
```

```python
list(previsoes)
```

```python
valores_previsoes = []
for p in regressor.predict(input_fn=funcao_previsao):
    valores_previsoes.append(p['predictions'])
```

```python
import numpy as np
valores_previsoes = np.asanyarray(valores_previsoes).reshape(-1, 1)
```
